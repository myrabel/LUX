```python
movies_per_year = newdf['year'].value_counts().sort_index()
movies_per_year.head()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o')
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid()

# Adjust x-axis ticks to skip a few years
years = movies_per_year.index.tolist()
plt.xticks(years[::5])  # Display every 5th year on the x-axis
plt.xticks(rotation=90)  # Rotate the x-axis labels by 90 degrees
plt.tight_layout()
plt.show()
```

```python
#create df for organizing movie titles by rating
top_rated_movies = newdf.sort_values(by='rating', ascending=False)

#remove duplicates in title column
top_rated_movies = top_rated_movies.drop_duplicates(subset=['title'])
top_rated_movies.head(10)
```
