A little more cleanup where we convert the "duration"column into type int from type str. This ensures I can categorize the data further

```python
newdf['runtime'] = newdf['duration'].str.extract(r'(\d+)')
newdf['runtime'] = newdf['runtime'].fillna(0).astype(int)
newdf.head()
```

Creating 2 new categories: Movies and Series based on the runtime (minutes) anything running for time greater than 100 minutes is categorized as a movie

```python
def categorize_runtime(runtime):
    if runtime >= 100:
        return 'Movie'
    else:
        return 'Series'

# Apply the function to create a new column
newdf['category'] = df['runtime'].apply(categorize_runtime)
newdf.head(10)
```

```python
movies_df = newdf[newdf['category'] == 'Movie']
del movies_df['end_year']
del movies_df['duration']
movies_df.head(10)
```

```python
movies_per_year = movies_df['start_year'].value_counts().sort_index()
movies_per_year.head()
```

Visualize number of movie releases per year. The sharp increase happens from the year 2009

```python
plt.figure(figsize=(15, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o')
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid()

# Adjust x-axis ticks to skip a few years
years = movies_per_year.index.tolist()
plt.xticks(years[::2])  # Display every 5th year on the x-axis
plt.xticks(rotation=90)  # Rotate the x-axis labels by 90 degrees
plt.tight_layout()
plt.show()
```

```python
series_df = newdf[newdf['category'] == 'Series']
del series_df['end_year']
del series_df['duration']
series_df.head(10)
```

```python
series_per_year = series_df['start_year'].value_counts().sort_index()
series_per_year.head()
```

Visualize number of series released per year. We can see a spike in releases from the year 2000 onwards

```python
plt.figure(figsize=(15, 6))
plt.plot(series_per_year.index, series_per_year.values, marker='o')
plt.title('Number of Series Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Series')
plt.grid()

# Adjust x-axis ticks to skip a few years
years = series_per_year.index.tolist()
plt.xticks(years[::2])  # Display every 5th year on the x-axis
plt.xticks(rotation=90)  # Rotate the x-axis labels by 90 degrees
plt.tight_layout()
plt.show()
```

```python
#create df for organizing movie titles by rating
top_rated_movies = movies_df.sort_values(by='rating', ascending=False)

#remove duplicates in title column
# top_rated_movies = top_rated_movies.drop_duplicates(subset=['title'])
top_rated_movies.head(10)
```

Visualize the top rated movies and series

```python
top_rated_movies = top_rated_movies.head(20)
plt.figure(figsize=(10, 6))
plt.barh(top_rated_movies['title'], top_rated_movies['rating'])
plt.gca().invert_yaxis()  # Invert y-axis to have the highest-rated movie at the top
plt.title('Highest Rated Movies')
plt.xlabel('Rating')
plt.ylabel('Movie Title')
plt.show()
print(top_rated_movies[['title', 'rating']])
```
