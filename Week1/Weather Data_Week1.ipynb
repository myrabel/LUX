```python
#Drop rows with blanks
new_df = data.dropna()
new_df.head()
```

```python
#1.	Find all records where the weather was exactly clear.

df1 = new_df.loc[new_df['Weather'] == 'Clear']
display(df1)
```

```python
#2.	Find the number of times the wind speed was exactly 4 km/hr

df2 = new_df.loc[new_df['Wind Speed_km/h'] == 4.0]
display(df2)
```

```python
#3.	Check if there are any NULL values present in the dataset.
new_df.isnull()
```

```python
#4.	Rename the column "Weather" to "Weather_Condition."
new_df.rename(columns = {'Weather':'Weather_Condition'}, inplace = True)
new_df.head()
```

```python
#5.	What is the mean visibility of the dataset?
visibility = new_df["Visibility_km"].mean()
print(visibility)
```

```python
#6.	Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
df3 = new_df.loc[(new_df['Wind Speed_km/h'] > 24) & (new_df['Visibility_km'] == 25.0)]
display(df3)
count_row = df3.shape[0]  # Gives number of rows
```

```python
#7.	What is the mean value of each column for each weather condition?
temp = new_df["Temp_C"].mean()
dewpoint = new_df["Dew Point Temp_C"].mean()
humidity = new_df["Rel Hum_%"].mean()
wind = new_df["Wind Speed_km/h"].mean()
visibility = new_df["Visibility_km"].mean()
pressure = new_df["Press_kPa"].mean()


#display(temp, dewpoint, humidity, wind, visibility, pressure)
print("Temperature", temp, "\n" "DewPoint", dewpoint, "\n" "Humidity", humidity,"\n" "Wind Speed", wind,"\n" "Visibility", visibility,"\n" "Pressure", pressure)
```

```python
#8.	Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
df4 = new_df.loc[(new_df['Weather_Condition'] == "Clear") & (new_df['Rel Hum_%'] > 50)]
df5 =new_df.loc[(new_df['Weather_Condition'] == "Clear") & (new_df['Visibility_km'] > 40) ]

#concatenate both tables
df6 = pd.concat([df4, df5])
df6
```

```python
#9.	Find the number of weather conditions that include snow.
df7 = new_df.loc[new_df['Weather_Condition'] == 'Snow']
display(df7)
```
